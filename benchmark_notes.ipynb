```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Pojedynek = ["Głębokość 1", "Głębokość 2", "Głębokość 3", "Głębokość 4"]
alpha_beta = [(82.2, 18.669761648183943), (877.4, 283.93844403320946), (27360.0, 8934.98326803134), (310801.8, 54146.88582143945)]
time_h1, std_h1 = list(zip(*alpha_beta))

alpha_beta = [(78.4, 12.658593918757328), (1002.2, 200.50576051575177), (20707.6, 7622.946139124951), (311981.4, 233007.9737739462)]

time_h2, std_h2 = list(zip(*alpha_beta))

alpha_beta = [(94.0, 12.312595177297109), (1368.6, 359.1816253652182), (26167.8, 11285.56783507148), (355917.0, 395057.5656017741)]
time_h3, std_h3 = list(zip(*alpha_beta))


bar_width = 0.2
bar_positions = range(len(Pojedynek))

sns.set_palette("Set2")

# Plotting mean values
plt.bar(bar_positions, time_h1, width=bar_width, yerr=std_h1, label='Heurystyka 1', log='y')
plt.bar([pos + bar_width for pos in bar_positions], time_h2, width=bar_width, yerr=std_h2, label='Heurystyka 2', log='y')
plt.bar([pos + 2*bar_width for pos in bar_positions], time_h3, width=bar_width, yerr=std_h3, label='Heurystyka 3', log='y')


plt.title('Liczba odwiedzonych wierzchołków podczas wyszukiwania najlepszego ruchu w pozycji (skala logarytmiczna)')
plt.ylabel('Odwiedzone wierzchołki')
plt.xticks([pos + bar_width / 2 for pos in bar_positions], Pojedynek)
plt.legend(loc="upper left")

plt.show()
```
