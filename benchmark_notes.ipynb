```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


languages = ["Python", "C++"]
mean_values = [587809, 73196]
std_values = [17049107, 3495399]
median_values = [512400, 31900]

sns.set_style("whitegrid")
bar_width = 0.3
bar_positions = [0, 1]

sns.set_palette("Set2")
plt.bar(bar_positions, mean_values, width=bar_width, label='Średnia')
plt.bar([pos + bar_width for pos in bar_positions], median_values, width=bar_width, label='Mediana')

plt.title('Średnia i mediana')
plt.ylabel('Czas [ns]')
plt.xticks([pos + bar_width / 2 for pos in bar_positions], languages)
plt.legend()
```
